```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Se usar o arquivo descompactado
#pd.read_csv('DOM2015.csv',sep=',')

#Se usar o arquivo compactado como está no dropbox púbico
dados2015 = pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
# Divisão por regiões do Brasil
dados2015.loc[(dados2015.UF > 29)  (dados2015.UF <= 35),"Região"] = "Sudeste"
dados2015 = dados2015[pd.notnull(dados2015["UF"])]

# Divisão por disponibilidade de internet no domicílio
dados2015.loc[(dados2015.V02322 == 2), "Internet"] = "Sim"
dados2015.loc[(dados2015.V02322 == 4), "Internet"] = "Não"
dados2015 = dados2015[pd.notnull(dados2015["V02322"])]

# Divisão por disponibilidade de tv por assinatura nos domicílios
dados2015.loc[(dados2015.V02272 == 1), "TvAss"] = "Sim"
dados2015.loc[(dados2015.V02272 == 3), "TvAss"] = "Não"
dados2015 = dados2015[pd.notnull(dados2015["V02272"])]

# Divisão por zona (Rural ou Urbana)
dados2015.loc[(dados2015.V4105 == 1), "Zona"] = "Urbana"
dados2015.loc[(dados2015.V4105 == 2), "Zona"] = "Urbana"
dados2015.loc[(dados2015.V4105 == 3), "Zona"] = "Urbana"
dados2015.loc[(dados2015.V4105 == 4), "Zona"] = "Rural"
dados2015.loc[(dados2015.V4105 == 5), "Zona"] = "Rural"
dados2015.loc[(dados2015.V4105 == 6), "Zona"] = "Rural"
dados2015.loc[(dados2015.V4105 == 7), "Zona"] = "Rural"
dados2015.loc[(dados2015.V4105 == 8), "Zona"] = "Rural"

# Renda
dados2015 = dados2015.rename(columns={"V4614": "Renda"})

# Tabela Com todas as Colunas
dados2015.loc[:,["Região", "Zona", "Internet", "TvAss", "Renda"]]
```

```python
# Analises Primarias
dadosanalise = pd.crosstab(dados2015.Internet, dados2015.TvAss)
dadosanalise
```

```python
corte2015 = dados2015.Renda[dados2015.TvAss == "Sim"].min()
dados2015_corte = dados2015.loc[dados2015.Renda >= corte2015]
dados2015_corte.loc[:,["Região", "Zona", "Internet", "TvAss", "Renda"]]
```
